```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
```

```python
df.head(1000)
```

```python
np.shape(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns=['state','date','freq','estimated unemployment rate','Estimated Employed',
           'Estimated Labour Participation Rate','region','longitude','latitude']
```

```python
df
```

```python
df.corr(method='pearson', min_periods=1, numeric_only=True)
```

```python
sns.heatmap(df.corr(method='pearson', min_periods=1, numeric_only=True),annot=True)
```

```python
ump=df[['state','region','estimated unemployment rate']]
fgr=px.sunburst(ump,path=['region','state'],
                values='estimated unemployment rate')
fgr.show()
```

```python
plt.bar(df['region'],df['estimated unemployment rate'])
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Unemployement rate region wise')
plt.show()
```

```python

plt.bar(df['state'],df['estimated unemployment rate'])
plt.xlabel('State')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Unemployement rate State wise')
plt.show()
```

```python
fig = px.line_3d(df, x="region", y="state", z="estimated unemployment rate")
fig.show()
```
